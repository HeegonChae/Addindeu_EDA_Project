### 1. 노지 과수 재배면적 DB생성 

```python
import pandas as pd
import re
```

```python
open_field_data = pd.read_csv('../project_data/노지_사과_재배면적.csv')

open_field_data.head()
```

```python
pattern = re.compile('^[\d.]+$')

new_columns = []
for column in open_field_data.columns:
    if pattern.match(column):
        new_columns.append('year' + column)
    else:
        new_columns.append(column)

open_field_data.columns = new_columns

open_field_data.head()
```

### 노지 채소 데이터 프레임 형성 

```python
veji_area_rate = pd.read_csv("../project_data/노지_채소_재배면적_1990.csv",encoding='utf-8')
veji_area_rate.drop(columns=['종류별(1)'],inplace=True)
veji_area_rate.head()
```

```python
def alter_columns_name(dataframe):
    new_columns = []
    pattern = re.compile('^[\d.]+$')
    for column in dataframe.columns:
        if pattern.match(column):
            new_columns.append('year' + column)
        else:
            new_columns.append(column)  
    return new_columns
```

```python
veji_area_rate.columns = alter_columns_name(veji_area_rate)
veji_area_rate.head()
```

```python
from sqlalchemy import create_engine

host = "database-1.cdv9nppqt3kw.ap-northeast-2.rds.amazonaws.com"
port = 3306
user = "hyun"
password = "0774"
database ="AutoFarming"

engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}/{database}')

table_name = 'openfield_vege'

veji_area_rate.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

engine.dispose()
```
